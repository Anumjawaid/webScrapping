```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
name=[]
year=[]
time=[]
genre=[]
rating=[]
metascore=[]
director=[]
votes=[]
money=[]
```

```python
pages=np.arange(1,1001,50)
```

```python
for pin in pages:
    #pin=("https://www.imdb.com/search/title/?groups=top_1000&start="+str(pin)+"&ref_=adv_prv")
    pin=requests.get("https://www.imdb.com/search/title/?groups=top_1000&start="+str(pin)+"&ref_=adv_prv")
    soup=BeautifulSoup(pin.text,"html.parser")
    soup.prettify()
    result=soup.find_all('div',attrs={"class":"lister-item-content"})
    #take break buddy
    sleep(randint(3,10))
    
```

```python
for i,q in enumerate(result):
    #print()
    t=q.h3.a.text#title
    name.append(t)
    a=q.h3.find('span',attrs={"class":"lister-item-year text-muted unbold"}).text#year
    year.append(a)
    v=q.find('span',attrs={"class":"runtime"}).text#time
    time.append(v)
    e=q.find('span',attrs={"class":"genre"}).text[1:].strip()#genre
    genre.append(e)
    r=q.strong.text#rating
    rating.append(r)
    try:
        u=q.find('span',attrs={'class':"metascore favorable"}).text[0:2]#metascore
        metascore.append(u)
    except:
        metascore.append("-")
    s=q.find('p',attrs={"class":""}).a.text#director
    director.append(s)
#result[0].find(['a','p'],attrs={'class':''})#work on starcast
    d=q.find('span',attrs={"name":"nv"}).text#votes
    votes.append(d)
    try:
        #g=q.find('p',attrs={"class":"sort-num_votes-visible"}).text[25:-1]#paisay
        g=q.find('p',attrs={'class':"sort-num_votes-visible"}).find_all('span',attrs={'name':'nv'})[1].text[1:]

        money.append(g)
    except:
        
        money.append("-")
```

```python
len(money)
```

```python
df=pd.DataFrame({
    "NAME":name,
    "YEAR":year,
    "STARS":rating,
    "VOTES":votes,
    "MSCORE":metascore,
    "DIRECTOR":director,
    "GROSS($)":money,
    "DURATION(min)":time,
    "GENRE":genre
})
```
