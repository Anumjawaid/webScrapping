```python
records = []
for reslt in result:
    Course = reslt.find('h2').text[0:]
    Uni_Name = reslt.find('h3').text[0:]
    Start_Date = reslt.find('p').text[53:65]
    Duration = reslt.contents[7].text[40:46]
    records.append((Course,Uni_Name,Start_Date,Duration))
```

```python
import pandas as pd
```

```python
table = pd.DataFrame(records,columns =['Course','Uni_Name','Start_Date','Duration'])
```
